```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv(r'C:\Users\Moi\Documents\python\ProjetDataHS-main/top_hearthstone_decks_20200221.csv')
df
```

```python
df.sort_values('games', ascending=False)
```

```python
df1 = df.groupby(['type']).mean()
df1 = df1.drop(['games', 'wins'], axis = 1)
df1['wr'].sort_values(ascending = False)
df1
```

```python

df2 = df.groupby(['type']).sum()
df2 = df2[['games','wins']]
df2.sort_values('games',ascending = False)
```

```python
sns.lineplot(x="dust", y="wr",data=df1)
```

```python
plt.figure(figsize = (12,6))
sns.lineplot(x = df2['games'] , y = df1['wr'])
```

```python
plt.figure(figsize = (12,6))
df3 = df2[(df2['games']< 100000)]
sns.lineplot(x = df3['games'] , y = df1['wr'])
```

```python
plt.figure(figsize = (12,6))
df4 = df.groupby("hero").count()
sns.barplot( x = df4.index, y = "type", data = df4)
```

```python
df5 = df.sort_values('duration', ascending = False)
df5 = df[(df['wr']> 0.50)]
df5 = df[(df['duration']< 6)]
sns.lineplot(x = 'duration', y = 'wr',data = df5)
```

```python
df6 = df
df6['wr/duration'] = df6["wr"]/df6["duration"] 
df6.sort_values("wr/duration", ascending = False)
sns.lineplot(x = 'hero', y = 'wr/duration',data = df6)
```

```python
plt.figure(figsize = (16,6))

df7 = df[(df['hero'] == "Hunter")]
df7['wr/duration'] = df7["wr"]/df7["duration"] 
df7.sort_values("wr/duration", ascending = False)
sns.barplot(x = 'type', y = 'wr/duration',data = df7)
```
